```python
# A Python Program to Solve a Sumaddle

from itertools import combinations, permutations


# Check patterns of each row

def check_zero_zero(row):    
    for i in range(1, len(row)):
        if row[i] == 0 and row[i-1] == 0:
            return True
    return False

def check_zero_x_zero(row):    
    for i in range(2, len(row)):
        if row[i-2] == 0 and row[i-1] > 0 and row[i] == 0:
            return True
    return False

def check_zero_x_x_zero(row):
    first_zero = -1
    for i in range(len(row)):
        if row[i] == 0:
            if first_zero < 0:
                first_zero = i
            else:
                if i > (found_first + 1):
                    return True
    return False

# Check row or column sum constraint
def check_blocksum(row, s):
    n = len(row)
    if s == 0:
        return check_zero_zero(row)
    else:
        blocksum = 0
        found = False
        for i in range(n):
            if not found:
                if row[i] == 0:
                    found = True
            elif row[i] > 0:
                blocksum +=row[i]
            else:
                break
        return s == blocksum
                   
def generate_candidates(n):
    num_ranges = [0] + [ i for i in range(n-1)]
    num_perms = [ list(val) for val in permutations(num_ranges, n)]
    candidates_separate_zeros= []
    candidates_consecutive_zeros = []
    for row in num_perms:
        if check_zero_zero(row):
            candidates_consecutive_zeros.append(row)
        else:
            candidates_separate_zeros.append(row)
    return candidates_separate_zeros, candidates_consecutive_zeros

def get_num_list(n, exclude_list):
    if exclude_list:
        num_of_zeros = exclude_list.count(0)        
        if num_of_zeros == 1:
            num_list = [0] + [i for i in range(1, n-1) if i not in exclude_list]
        elif num_of_zeros == 2:
             num_list =  [i for i in range(1, n-1) if i not in exclude_list]
        else:
            num_list = [0] +  [i for i in range(1, n-1) if i not in exclude_list]   
    else:
        num_list = [ i for i in range(n-1)]
    return sorted(num_list)

def get_last_num(n, exclude_list):    
    num_list =  [i for i in range(n-1) if i not in exclude_list]
    if len(num_list) == 0:
        return 0
    else:
        return num_list[0]
  

def generate_last_row(n,row_sum, prefix):
    row = []
    for i in range(n):
        exclude_numbers = [row[i] for row in prefix]
        last_num = get_last_num(n, exclude_numbers)
        row.append(last_num)
    if not check_blocksum(row, row_sum):
        return None
    else:
        return row

def check_prefix_constraint(n,row, prefix):
    """
    Check if current row has a duplicate value in prefix
    """    
    for i in range(n):
        exclude_numbers =  [r[i] for r in prefix]
        if row[i] > 0:
            if row[i] in exclude_numbers:
                return False
        else:
            if exclude_numbers.count(0) == 2:
                return False
    return True
        
    
def generate_permutations_v0(n, prefix):
    """
    Generate permutations for [0, 0, 1, 2, ... n-2] with a constraint given by a prefix
    """
    num_list_full = [0] + [ i for i in range(n-1)]
    perms = [list(val) for val in permutations(num_list_full, n)] 
    if len(prefix) and len(prefix[0]):
        perms_constrained = []
        for row in perms:
            if check_prefix_constraint(n, row, prefix):
                perms_constrained.append(row)
        return perms_constrained
    else:
        return perms
    
    
def generate_permutations(n, prefix):
    
    if len(prefix) and len(prefix[0]):
        num_perms = []
        num_list = [[] for i in range(n)]        
        exclude_numbers = [row[0] for row in prefix] 
        num_list0  = get_num_list(n, exclude_numbers)
        
        if n == 4:
            for r0 in num_list0:                
                if r0 > 0:
                    exclude_numbers =  [row[1] for row in prefix] + [r0]
                else:
                    exclude_numbers =  [row[1] for row in prefix]
                num_list1 = get_num_list(n, exclude_numbers)
                for r1 in num_list1:                 
                    parent_list = [r0, r1]           
                    exclude_numbers =  [row[2] for row in prefix] + [k for k in parent_list if k > 0]
                    if parent_list.count(0) > 1:
                        exclude_numbers.extend([0,0])                               
                    num_list2 = get_num_list(n, exclude_numbers) 
                    for r2 in num_list2:
                        exclude_numbers = [r0, r1, r2]
                        r3 = get_last_num(n, exclude_numbers)
                        num_perms.append([r0, r1, r2, r3])                            
        elif n == 5:      
            for r0 in num_list0:
                if r0 > 0:
                    exclude_numbers =  [row[1] for row in prefix] + [r0]
                else:
                    exclude_numbers =  [row[1] for row in prefix]
                num_list1 = get_num_list(n, exclude_numbers)
                for r1 in num_list1:                 
                    parent_list = [r0, r1]           
                    exclude_numbers =  [row[2] for row in prefix] + [k for k in parent_list if k > 0]
                    if parent_list.count(0) > 1:
                        exclude_numbers.extend([0,0])                               
                    num_list2 = get_num_list(n, exclude_numbers) 
                    for r2 in num_list2:
                        parent_list = [r0, r1, r2]
                        exclude_numbers =  [row[3] for row in prefix] + [k for k in parent_list if k > 0]                 
                        if parent_list.count(0) > 1:
                            exclude_numbers.extend([0,0])                            
                        num_list3 = get_num_list(n, exclude_numbers)  
                        for r3 in num_list3:
                            exclude_numbers = [r0, r1, r2, r3]
                            r4 = get_last_num(n, exclude_numbers)
                            num_perms.append([r0, r1, r2, r3,r4])
        elif n == 6:      
            for r0 in num_list0:
                if r0 > 0:
                    exclude_numbers =  [row[1] for row in prefix] + [r0]
                else:
                    exclude_numbers =  [row[1] for row in prefix]
                num_list1 = get_num_list(n, exclude_numbers)
                for r1 in num_list1:                 
                    parent_list = [r0, r1]           
                    exclude_numbers =  [row[2] for row in prefix] + [k for k in parent_list if k > 0]
                    if parent_list.count(0) > 1:
                        exclude_numbers.extend([0,0])                               
                    num_list2 = get_num_list(n, exclude_numbers) 
                    for r2 in num_list2:
                        parent_list = [r0, r1, r2]
                        exclude_numbers =  [row[3] for row in prefix] + [k for k in parent_list if k > 0]                 
                        if parent_list.count(0) > 1:
                            exclude_numbers.extend([0,0])                            
                        num_list3 = get_num_list(n, exclude_numbers)  
                        for r3 in num_list3:
                            parent_list = [r0, r1, r2, r3]
                            exclude_numbers =  [row[4] for row in prefix] + [k for k in parent_list if k > 0]                                                
                            if parent_list.count(0) > 1:
                                exclude_numbers.extend([0,0])                    
                            num_list4 = get_num_list(n, exclude_numbers)                          
                            for r4 in num_list4:
                                exclude_numbers = [r0,r1,r2,r3,r4]
                                r5 = get_last_num(n, exclude_numbers)
                                num_perms.append([r0,r1,r2,r3,r4,r5])
        elif n == 7:      
            for r0 in num_list0:
                if r0 > 0:
                    exclude_numbers =  [row[1] for row in prefix] + [r0]
                else:
                    exclude_numbers =  [row[1] for row in prefix]
                num_list1 = get_num_list(n, exclude_numbers)
                for r1 in num_list1:                 
                    parent_list = [r0, r1]           
                    exclude_numbers =  [row[2] for row in prefix] + [k for k in parent_list if k > 0]
                    if parent_list.count(0) > 1:
                        exclude_numbers.extend([0,0])                               
                    num_list2 = get_num_list(n, exclude_numbers) 
                    for r2 in num_list2:
                        parent_list = [r0, r1, r2]
                        exclude_numbers =  [row[3] for row in prefix] + [k for k in parent_list if k > 0]                 
                        if parent_list.count(0) > 1:
                            exclude_numbers.extend([0,0])                            
                        num_list3 = get_num_list(n, exclude_numbers)  
                        for r3 in num_list3:
                            parent_list = [r0, r1, r2, r3]
                            exclude_numbers =  [row[4] for row in prefix] + [k for k in parent_list if k > 0]                                                
                            if parent_list.count(0) > 1:
                                exclude_numbers.extend([0,0])                    
                            num_list4 = get_num_list(n, exclude_numbers)                               
                            for r4 in num_list4:
                                parent_list = [r0, r1, r2, r3, r4]
                                exclude_numbers =  [row[5] for row in prefix] + [k for k in parent_list if k > 0]                       
                                if parent_list.count(0) > 1:
                                    exclude_numbers.extend([0,0])                            
                                num_list5 = get_num_list(n, exclude_numbers)                          
                                for r5 in num_list5:
                                    exclude_numbers = [r0,r1,r2,r3,r4,r5]
                                    r6 = get_last_num(n, exclude_numbers)
                                    num_perms.append([r0,r1,r2,r3,r4,r5,r6])                            
                                                     
                            
    else:
        num_list_full = [0] + [ i for i in range(n-1)]
        num_perms = [ list(val) for val in permutations(num_list_full, n)]
    
    return num_perms

def generate_row_candidates_new(n,k, row_sum, prev_rows, prev_cols):
    candidates = generate_permutations(n-k,prev_rows, prev_cols)
    # Filter row candidates
    if row_sum == 0:
        candidates = [row for row in candidates if check_zero_zero(row)]
    elif row_sum > 0: 
        candidates = [row for row in candidates if check_blocksum(row, row_sum)]          
        # Check patterns - courtesy of Brian Mansfield
        if row_sum in (1, 2):
            candidates = [row for row in candidates if check_zero_x_zero(row)]        
    return candidates

def generate_row_candidates(n,row_sum, prefix):
    candidates = generate_permutations(n, prefix) 
    # Filter row candidates
    if row_sum == 0:
        candidates = [row for row in candidates if check_zero_zero(row)]
    elif row_sum > 0: 
        candidates = [row for row in candidates if check_blocksum(row, row_sum)]          
        # Check patterns - courtesy of Brian Mansfield
        if row_sum in (1, 2):
            candidates = [row for row in candidates if check_zero_x_zero(row)]        
    return candidates

def check_col(col):
    seen = {}
    for val in col:
        if val in seen.keys():
            if val > 0:
                return False
            elif seen[val] > 1:
                return False
            else:
                seen[val] +=1
        else:
            seen[val]  = 1
    return True


def check_col_constraint(sol, col_sums, n): 
    nrow = len(sol)
    for i in range(n):
        col = [row[i] for row in sol]
        if not check_col(col):
            return False
        if col_sums[i] == 0:
            if not check_zero_zero(col):
                return False
        elif col_sums[i] > 0:
            if not check_blocksum(col, col_sums[i]):
                return False
    return True

# Generate solutions for 5x5, 6x6 or 7x7
def generate_solutions(row_sums,col_sums, n):
    if n == 4:
        row_candidates0 = generate_row_candidates(n,row_sums[0], [[]])
        for r0 in row_candidates0:
            row_candidates1 = generate_row_candidates(n,row_sums[1], [r0]) 
            for r1 in row_candidates1:  
                row_candidates2 = generate_row_candidates(n,row_sums[2], [r0,r1])             
                for r2 in row_candidates2:  
                    r3 = generate_last_row(n,row_sums[4], [r0,r1,r2]) 
                    if r3:  
                        sol = [r0,r1,r2,r3] 
                        if check_col_constraint(sol, col_sums, n): 
                            return sol
    elif n == 5:
        row_candidates0 = generate_row_candidates(n,row_sums[0], [[]])
        for r0 in row_candidates0:
            row_candidates1 = generate_row_candidates(n,row_sums[1], [r0]) 
            for r1 in row_candidates1:  
                row_candidates2 = generate_row_candidates(n,row_sums[2], [r0,r1])             
                for r2 in row_candidates2:
                    row_candidates3 = generate_row_candidates(n,row_sums[3], [r0,r1,r2])  
                    for r3 in row_candidates3:  
                        r4 = generate_last_row(n,row_sums[4], [r0,r1,r2,r3]) 
                        if r4:  
                            sol = [r0,r1,r2,r3,r4] 
                            if check_col_constraint(sol, col_sums, n): 
                                return sol
    elif n == 6:
        row_candidates0 = generate_row_candidates(n,row_sums[0], [[]])
        for r0 in row_candidates0:
            row_candidates1 = generate_row_candidates(n,row_sums[1], [r0]) 
            for r1 in row_candidates1:  
                row_candidates2 = generate_row_candidates(n,row_sums[2], [r0,r1])             
                for r2 in row_candidates2:
                    row_candidates3 = generate_row_candidates(n,row_sums[3], [r0,r1,r2])  
                    for r3 in row_candidates3:
                        row_candidates4 = generate_row_candidates(n,row_sums[4], [r0,r1,r2,r3])                      
                        for r4 in row_candidates4:  
                            r5 = generate_last_row(n,row_sums[5], [r0,r1,r2,r3,r4]) 
                            if r5:  
                                sol = [r0,r1,r2,r3,r4,r5] 
                                if check_col_constraint(sol, col_sums, n): 
                                    return sol                              
    elif n == 7:
        row_candidates0 = generate_row_candidates(n,row_sums[0], [[]])
        for r0 in row_candidates0:
            row_candidates1 = generate_row_candidates(n,row_sums[1], [r0] ) 
            for r1 in row_candidates1:  
                row_candidates2 = generate_row_candidates(n,row_sums[2], [r0, r1])             
                for r2 in row_candidates2:
                    row_candidates3 = generate_row_candidates(n,row_sums[3], [r0,r1,r2])  
                    for r3 in row_candidates3:
                        row_candidates4 = generate_row_candidates(n,row_sums[4], [r0,r1,r2,r3])                      
                        for r4 in row_candidates4:  
                            row_candidates5 = generate_row_candidates(n,row_sums[5], [r0,r1,r2,r3,r4])               
                            for r5 in row_candidates5:  
                                r6 = generate_last_row(n,row_sums[6], [r0,r1,r2,r3,r4,r5]) 
                                if r6:  
                                    sol = [r0,r1,r2,r3,r4,r5,r6] 
                                    if check_col_constraint(sol, col_sums, n): 
                                        return sol    
        
    else:
        raise Exception("Sorry, the input n must be 5, 6, or 7")
    return None

# Generate solutions for 5x5, 6x6 or 7x7
def generate_solutions_v2(row_sums,col_sums, n):
    if n == 4:
        row_candidates0 = generate_row_candidates(n,row_sums[0], [[]])
        for r0 in row_candidates0:
            row_candidates1 = generate_row_candidates(n,row_sums[1], [r0]) 
            for r1 in row_candidates1:  
                row_candidates2 = generate_row_candidates(n,row_sums[2], [r0,r1])             
                for r2 in row_candidates2:  
                    r3 = generate_last_row(n,row_sums[4], [r0,r1,r2]) 
                    if r3:  
                        sol = [r0,r1,r2,r3] 
                        if check_col_constraint(sol, col_sums, n): 
                            return sol
    elif n == 5:
        row0 = generate_row_candidates(n,0,row_sums[0], [[]], [[]])
        for r0 in row0:
            col0 = generate_col_candidates(n,0,col_sums[0],[r0],[[]])
            for c0 in col0:  
                row1 = generate_row_candidates(n,1,row_sums[1], [r0], [c0]) 
                    for r1 in row1:  
                        col1 = generate_col_candidates(n,1,col_sums[1],[r0,r1],[c0])
                        for c1 in col1:                          
                            row2 = generate_row_candidates(n,2,row_sums[2], [r0,r1], [c0, c1])                                       
                            for r2 in row2:
                                col2 = generate_col_candidates(n,2,col_sums[2],[r0,r1,r2],[c0,c1])
                                for c2 in col2:                                      
                                    row3 = generate_row_candidates(n,row_sums[3], [r0,r1,r2])  
                                    for r3 in row3:  
                                        r4 = generate_last_row(n,row_sums[4], [r0,r1,r2,r3]) 
                                        if r4:  
                                            sol = [r0,r1,r2,r3,r4] 
                                            if check_col_constraint(sol, col_sums, n): 
                                                return sol
    elif n == 6:
        row_candidates0 = generate_row_candidates(n,row_sums[0], [[]])
        for r0 in row_candidates0:
            row_candidates1 = generate_row_candidates(n,row_sums[1], [r0]) 
            for r1 in row_candidates1:  
                row_candidates2 = generate_row_candidates(n,row_sums[2], [r0,r1])             
                for r2 in row_candidates2:
                    row_candidates3 = generate_row_candidates(n,row_sums[3], [r0,r1,r2])  
                    for r3 in row_candidates3:
                        row_candidates4 = generate_row_candidates(n,row_sums[4], [r0,r1,r2,r3])                      
                        for r4 in row_candidates4:  
                            r5 = generate_last_row(n,row_sums[5], [r0,r1,r2,r3,r4]) 
                            if r5:  
                                sol = [r0,r1,r2,r3,r4,r5] 
                                if check_col_constraint(sol, col_sums, n): 
                                    return sol                              
    elif n == 7:
        row_candidates0 = generate_row_candidates(n,row_sums[0], [[]])
        for r0 in row_candidates0:
            row_candidates1 = generate_row_candidates(n,row_sums[1], [r0] ) 
            for r1 in row_candidates1:  
                row_candidates2 = generate_row_candidates(n,row_sums[2], [r0, r1])             
                for r2 in row_candidates2:
                    row_candidates3 = generate_row_candidates(n,row_sums[3], [r0,r1,r2])  
                    for r3 in row_candidates3:
                        row_candidates4 = generate_row_candidates(n,row_sums[4], [r0,r1,r2,r3])                      
                        for r4 in row_candidates4:  
                            row_candidates5 = generate_row_candidates(n,row_sums[5], [r0,r1,r2,r3,r4])               
                            for r5 in row_candidates5:  
                                r6 = generate_last_row(n,row_sums[6], [r0,r1,r2,r3,r4,r5]) 
                                if r6:  
                                    sol = [r0,r1,r2,r3,r4,r5,r6] 
                                    if check_col_constraint(sol, col_sums, n): 
                                        return sol    
        
    else:
        raise Exception("Sorry, the input n must be 5, 6, or 7")
    return None
   
def print_matrix(sol):
    for col in sol:
        print(col)
        
def solve_sumaddle_puzzle(row_sums, col_sums):
    n = len(row_sums)
    sol = generate_solutions(row_sums,col_sums, n)    
    if sol:
        print(f"Solution for a {n}x{n} with row constraint:{row_sums}, and column constraint:{col_sums}")
        print_matrix(sol)
    else:
        print("No solution")    
```
