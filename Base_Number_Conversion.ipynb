# Use Python to Convert Base Numbers

## Decimal to Binary

```python
# Decimal to Binary
n = 2022
bi = bin(n)
print("Decimal number:", n)
print("Binary number:", bi)
```

```python
# Program to remove leading binary prefix "0b"
s=bi[2:]
print(s)
```

## Binary to Decimal

```python
# Binary to Decimal
bi = "11111100110"
n = int(bi, 2)
print("Binary:", bi)
print("Decimal:", n)
```

## Decimal to Octal

```python
n = 2022
oc = oct(n)
print("Decimal number:", n)
print("Octal number:", oc)
```

```python
# remove leading prefix of "0o"
s=oc[2:]
print(s)
```

## Octal to Decimal

```python
# Oct to Decimal
s = "3746"
n = int(s, 8)
print("Octal:", s)
print("Decimal:", n)
```

## Decimal to Hex

```python
n = 2022
h = hex(n)
print("Decimal number:", n)
print("Hex number:", h)
```

```python
## remove leading prefix of "0x"
s=h[2:]
print(s)
```

## Hex to Decimal

```python
# Hex to Decimal
s = "7e6"
n = int(s, 16)
print("Hex:", s)
print("Decimal:", n)
```

## First 100 Numbers in Decimal, Binary, Octal and Hexadecimal

```python
print("Decimal, Binary, Octal, Hexadecimal")
for i in range(1,101):
    print(i, bin(i)[2:], oct(i)[2:], hex(i)[2:])
```

## Format a table

```python
import pandas as pd
pd.set_option('display.max_rows', None)
cols = ['Decimal', 'Binary', 'Octal', 'Hex']
list100=[[i, bin(i)[2:], oct(i)[2:], hex(i)[2:]] for i in range(100)]
df=pd.DataFrame(list100, columns=cols)
```
